```python
import pathlib

train_data_dir = 'cat_dog_photos/train' 
train_data_dir = pathlib.Path(train_data_dir)

test_data_dir = 'cat_dog_photos/test1'  
test_data_dir = pathlib.Path(test_data_dir)

train_data_dir, test_data_dir  
```

```python
# to get all train image paths 
train_paths = list(train_data_dir.glob('*.jpg')) 
train_paths, len(train_paths)  
```

```python
# to get all test image paths 
test_paths = list(test_data_dir.glob('*.jpg')) 
test_paths, len(test_paths) 
```

```python
# to categorize images by their labels
cats = list(train_data_dir.glob('cat*'))
dogs = list(train_data_dir.glob('dog*')) 
```

```python
# cat image
Image.open(str(cats[0])).resize((224,224))  
```

```python
# dog image 
Image.open(str(dogs[0])).resize((224,224)) 
```

```python
# cat and dog dictionary
cat_dog_dict = {
    'cats': cats,
    'dogs': dogs
}
```

```python
# cat image
Image.open(str(cat_dog_dict['cats'][3])).resize((224,224)) 
```

```python
# dog image
Image.open(str(cat_dog_dict['dogs'][3])).resize((224,224)) 
```

```python
# cat dog labels dictionary
cat_dog_labels_dict = {
    'cats': 0,
    'dogs': 1
}
```

```python
# get the size of an image 
img = cv2.imread(cat_dog_dict['cats'][0])
img.shape 
```

```python
# to convert all images to same size (180,180,3)
X, y = [], [] 

# add resized images to X and labels to y
for animal, images in cat_dog_dict.items(): 
    for image in images: 
        img = cv2.imread(str(image))
        resized_image = cv2.resize(img, (180,180)) 
        X.append(resized_image)
        y.append(cat_dog_labels_dict[animal]) 
```

```python
# convert X and y to numpy arrays 
X = np.array(X)
y = np.array(y) 
```

```python
# normalize the X values 
X_scaled = X / 255 
X_scaled[0] 
```

```python
# train test split 
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X_scaled,  y, train_size=0.2, random_state=32)

array = [X_train, X_test, y_train, y_test] 

for i in array: 
    print(len(i)) 
```
